```python
import cv2
from matplotlib import pyplot as plt


# Opening image/reading
img = cv2.imread(r"C:\Users\erami\OneDrive\Desktop\OpenCV\photo.jpeg")

# OpenCV opens images as BRG
# but we want it as RGB and
# we also need a grayscale
# version
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # this will convert color to gray scale BGR2GRAY
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)# this will convert gray scale to color image BGR2RGB
# bgr blue+green+red  # rgb red+green+blue
# Creates the environment
# of the picture and shows it
plt.subplot(1, 1, 1) # 
plt.imshow(img_rgb)
plt.show()
# subplot function takes 3 argument  that describe of layout of image
# row 1 , height , column 1 , 1

#subplot(1,2,2)
# row 1 , 2 column  # second plot
```
